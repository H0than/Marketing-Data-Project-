```python
#Import Packages 

import pandas as pd
```

```python
#Read in the CSV file

file_path = "marketing_campaign_data_py.csv"
df = pd.read_csv(file_path)
print ("CSV File has been successfully read in")
```

```python
#Verify the dataframe 

print (df)
df.head()
```

```python
#Rename few columns

df.rename (columns = {'Campaign':'Platform', 'Date':'Day', 'Cost':'Spend'}, inplace =True)
print (df.columns)
print ('Column names have been renamed')
df.head()
```

```python
#Converting Revenue and Spend into numeric values

df['Revenue'] = df['Revenue'].replace('[\$,]', '', regex=True).astype(float)
df['Spend'] = df['Spend'].replace('[\$,]', '', regex=True).astype(float)
df.head()
```

```python
#Calculate new metrics 

df['CTR'] = df['Clicks']/df['Impressions']
print ('CTR column has been created')


df['CPA'] = df['Spend']/df['Conversions']
print ('CPA column has been created')


df['CPC'] = df['Clicks']/df['Spend']
print ('CPC column has been created')


df['ROAS'] = df['Revenue']/df['Spend']
print ('ROAS column has been created')


df['CPM'] = (df['Spend']/df['Impressions']) * 100
print ('CPM column has been created')

df.head()
```

```python
# Round CTR, CPC, CPM, and ROAS columns to two decimal places

df['CTR'] = df['CTR'].round(2)
df['CPA'] = df['CPA'].round(2)
df['CPC'] = df['CPC'].round(2)
df['CPM'] = df['CPM'].round(2)
df['ROAS'] = df['ROAS'].round(2)

df.head()
```

```python
# Format 'CPC', 'CPA', 'ROAS','CPM', 'Spend' and 'Revenue' columns as currency

df['Spend'] = df['Spend'].map('${:,.2f}'.format)
df['Revenue'] = df['Revenue'].map('${:,.2f}'.format)
df['CPC'] = df['CPC'].map('${:,.2f}'.format)
df['CPA'] = df['CPA'].map('${:,.2f}'.format)
df['ROAS'] = df['ROAS'].map('${:,.2f}'.format)
df['CPM'] = df['CPM'].map('${:,.2f}'.format)

df.head()
```

```python
#Concatenate 'Platform' and 'Campaign Type' to create 'Campaign Name'

df['Campaign Name'] = df['Campaign Type'] + '-' + df['Platform']
df.head()
```

```python
# Convert CTR to percentage format

df['CTR'] = (df['CTR'] * 100).round(2).astype(str) + '%'
df.head()
```

```python
#Reorder columns
 
df = df [['Day', 'Platform','Campaign Type', 'Campaign Name','Impressions', 'Clicks',
       'Conversions', 'Revenue', 'Spend', 'Region', 'CTR','CPA', 'CPC','ROAS','CPM','AdCreative']]
print(df.columns)
```

```python
# Save the DataFrame to a CSV file

file_path = "Modified_Marketing_Campaign_Data_py.csv"

df.to_csv(file_path, index=False)

print("CSV file saved successfully.")
```
